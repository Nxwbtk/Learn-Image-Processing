Open CV Method

```python
import cv2

image2 = cv2.imread('img1.jpg')
```

```python
plt.imshow(image2)
plt.axis('off')
plt.show()
# Color in OpenCV is BGR not RGB so the image will be shown in wrong colors
# To fix this we can convert the image to RGB
image_convert = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.imshow(image_convert)
plt.axis('off')
plt.show()

# Or just reverse the order of the channels
image_reverse = image2[:, :, ::-1]
plt.imshow(image_reverse)
plt.axis('off')
plt.show()
```
